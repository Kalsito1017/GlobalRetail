```python
filePath = "/Workspace/Users/mygmail@gmail.com/GlobalRetail/Bronze_Layer/customer_data/customer.csv"
df = spark.read.csv(filePath, header=True, inferSchema=True)
df.show()
```

```python
from pyspark.sql.functions import current_timestamp
df_new = df.withColumn("ingestion_timestamp", current_timestamp())
df_new.show()
df_new.write.mode("overwrite").saveAsTable("customer_data")
display(df_new)
```

```python
spark.sql("use globalretail_bronze")
df_new.write.format("delta").mode("overwrite").saveAsTable("bronze_customer")
```
