```python
filePath = "/Workspace/Users/mygmail@gmail.com/GlobalRetail/Bronze_Layer/transaction/transaction.snappy.parquet"
df = spark.read.parquet(filePath)
df.display()
```

```python
from pyspark.sql.functions import to_timestamp, col

df = df.withColumn("transaction_date", to_timestamp(col("transaction_date")))
display(df)
df.printSchema()
```

```python
from pyspark.sql.functions import current_timestamp
final_df = df.withColumn("ingestion_timestamp", current_timestamp())
display(final_df)
```

```python
spark.sql("use globalretail_bronze")
final_df.write.format("delta").mode("append").saveAsTable("bronze_transaction")
```
