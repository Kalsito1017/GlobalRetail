```python
spark.sql("USE globalretail_gold")

spark.sql("""
CREATE OR REPLACE TABLE gold_daily_sales AS
SELECT
  transaction_date,
  SUM(total_amount) AS daily_total_sales
FROM
  globalretail_silver.silver_orders
GROUP BY
  transaction_date
""")
```

```python
df = spark.table("gold_daily_sales")
display(df)
```
